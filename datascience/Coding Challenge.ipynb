### 1. Task
Change the date format to 'YYYY-MM-DD' and select the data from 2005 till 2009 for station numbers including and between 725300 and 726300 , and save it as a pandas dataframe. Note the maximum year available is 2010. 

```python
%%bigquery
SELECT 
PARSE_DATE('%Y-%m-%d', concat(year, "-", month, "-", day)) AS DATE,
* EXCEPT (year, month, day), 
FROM `bigquery-public-data.samples.gsod`
WHERE year < 2010 AND year > 2004 AND station_number >= 725300 AND station_number <= 726300
LIMIT 20;
```

```python
import pandas as pd
from google.cloud import bigquery

client = bigquery.Client()
sql = """
SELECT 
PARSE_DATE('%Y-%m-%d', concat(year, "-", month, "-", day)) AS DATE,
* EXCEPT (year, month, day), 
FROM `bigquery-public-data.samples.gsod`
WHERE year < 2010 AND year > 2004 AND station_number >= 725300 AND station_number <= 726300;
"""

df = client.query(sql).to_dataframe()
```

```python
# drop columns that contain information about how the data points were collected
# (could also use these, e.g. by dropping rows where the temperature is only collected
# from few samples)
clean_df = task_df.drop(columns = ['num_mean_temp_samples', 'num_mean_dew_point_samples',
              'num_mean_sealevel_pressure_samples', 'num_mean_station_pressure_samples',
             'num_mean_visibility_samples', 'num_mean_wind_speed_samples',
             'max_temperature_explicit', 'min_temperature_explicit'])
```

```python
# drop columns that contain a lot of NaNs
clean_df = clean_df.drop(columns = ['mean_station_pressure', 'min_temperature',
                                    'snow_depth', 'mean_sealevel_pressure'])
```

```python
# replace booleans
clean_df = clean_df.replace(to_replace=[True, False], value=[1,0])
```

```python
# drop total precipitation, since that gives away whether it rained
clean_df = clean_df.drop(columns = ['total_precipitation'])
```

```python
# replace remaining NaNs with median values
clean_df = clean_df.fillna(clean_df.median())
```

### 3. Task
Now it is time to split the data, into a training, evaluation and test set. As a reminder, the date we are trying to predict snow fall for is the following, and hence should constitute your test set.

```python
import datetime as dt

str(dt.datetime.today()- dt.timedelta(days=13*365)).split(' ')[0]
```

This formula ignores leap days, so I'm going to use 2009-07-05.

```python
test_df = clean_df[clean_df.DATE == dt.date(2009,7,5)]
```

```python
test_df
```

```python
remaining = clean_df[clean_df.DATE != dt.date(2009,7,5)]
remaining = remaining.drop(columns = ['DATE', 'station_number', 'wban_number'])
train_df = remaining.sample(frac = 0.9)
eval_df = remaining.drop(train_df.index)
print(len(train_df[train_df.rain == True]))
print(len(eval_df[eval_df.rain == True]))
```
